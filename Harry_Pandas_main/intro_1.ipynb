HOW TO MERGE
---->

```python
df1 = pd.DataFrame({
    "Name": ["Anshu", "Ravi", "Sneha", "Priya"],
    "Marks": [85, 90, 78, 92]
})

df1
```

```python
df2 = pd.DataFrame({
    "Name": ["Amit", "Zaid", "Rohan", "Simran"],
    "Marks": [88, 76, 95, 89]
})

df2
```

```python
#---concat DataFrames---
pd.concat([df1, df2], axis=0)  # Concatenate DataFrames vertically (row-wise)
```

```python
#---concat DataFrames---
pd.concat([df1, df2], axis=1)  # Concatenate DataFrames horizontally (column-wise)
```

```python
df1 = pd.DataFrame({
    "Name": ["Anshu", "Aditi", "Anushka", "mariya", "Yash"],
    "Marks": [85, 90, 78, 92, 88]
})

df1
```

```python
df2 = pd.DataFrame({
    "Name": ["Anshu", "Aditi", "Anushka", "mariya", "Harry", "Yash"],
    "Roll No.": [1, 2, 3, 4, 9, 5]
})

df2
```

```python
#----now murge DataFrames---
pd.merge(df1, df2, on="Name")  # Merge DataFrames on the "Name" column
```

```python
#----now murge DataFrames---
df3=pd.merge(df1, df2, on="Name")  # Merge DataFrames on the "Name" column

df3
```

```python
## -----new file gnerate------
df3.to_csv("data/new_merge.csv", index=False)  # Save DataFrame to a new CSV file with no index colum before Id
```
